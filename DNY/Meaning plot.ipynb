```python
"""
Kindly use only to change parameters. Please enter all your info to dny.txt
"""

import matplotlib
from matplotlib import pyplot as plt, patches

def dtc(k):
    date = k.split('/') #date = yy/mm
    return (int(date[0]) - 16) + int(date[1])/12

plt.rcParams["figure.figsize"] = [16.00, 10.00]
plt.rcParams["figure.dpi"] = 80 #Increase resolution
plt.rcParams["figure.autolayout"] = True
fig = plt.figure()
ax = fig.add_subplot(111)

col = ['springgreen', 'lightsteelblue', 'navajowhite', 'mistyrose'] #Add colours to get more colours

file = open('dny.txt', 'r')
lines = file.read().split('\n')
#print(lines)
for x in range(1, len(lines)):
    wline = lines[x].split(',')
#    print(wline)
    color_cat, date, percent_meaning , text_in_circle, size = int(wline[0]), wline[1], float(wline[2]), wline[3], float(wline[4]) 
    # Category 1- 4, Date, Percent, Name, Importance 1-3
    circle = matplotlib.patches.Circle((dtc(date), percent_meaning*4), radius = size , color = col[color_cat - 1], alpha=0.25)
    ax.add_patch(circle)    
    plt.text(dtc(date)-0.015*len(text_in_circle), percent_meaning*4, text_in_circle)
file.close()
    
    
plt.yticks(range(5), ['0%', '25%', '50%', '75%', '100%'])
plt.xticks([1, 3, 5, 7], ['2017', '2018', '2019', '2020']) #Change years if you wish
plt.ylabel('Meaningfulness')
plt.xlabel('Years')
plt.xlim([0, 8])
plt.ylim([0, 5])
#plt.axis('equal')
plt.show()
```
